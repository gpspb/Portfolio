В датасете у нас всего 9 записей с аналогичным количеством гипотез, для каждой присвоены значения охвата пользователей(reach), предполагаемого влияния(impact), степени уверенности в присвоенных значениях(confidence) и величины трудозатрат на тестирование гипотезы(efforts) по шкале от одного до десяти.

Приоритизируем гипотезы с помощью фреймворков `ICE` и `RICE` и добавим эти показатели в датафрейм.

```python
df_h['ICE'] = (df_h['impact'] * df_h['confidence']) / df_h['efforts']
df_h[['hypothesis', 'ICE']].sort_values(by='ICE', ascending=False).style.format({('ICE'): "{:.2f}"})
```

Лидерами по `ICE` оказались гипотезы с индексами 8, 0, 7 и 6. Заметим, что при расчёте рейтинга `ICE` не учитывается охват аудитории, поэтому гипотезу под номером 8 не стоит рассматривать как самую срочную. К тому же, чтобы собрать две релевантные группы по количеству пользователей для данного теста может потребоваться очень много времени.

```python
df_h['RICE'] = (df_h['reach'] * df_h['impact'] * df_h['confidence']) / df_h['efforts']
df_h[['hypothesis', 'RICE']].sort_values(by='RICE', ascending=False).style.format({('RICE'): "{:.2f}"})
```

Лидерами по `RICE` оказались гипотезы с индексами 7, 2, 0 и 6. Таким образом, с учётом пересечения результатов двух фреймворков, предпочтение стоит отдать гипотезам №0:**Добавить два новых канала привлечения трафика, что позволит привлекать на 30% больше пользователей**, №7: **Добавить форму подписки на все основные страницы, чтобы собрать базу клиентов для email-рассылок** и №6: **Показать на главной странице баннеры с актуальными акциями и распродажами, чтобы увеличить конверсию**.

## 5. Анализ A/B-теста.

Начнём с подготовки данных. Чтобы построить графики, нужно собрать кумулятивные данные. Найдём число уникальных ID заказов и пользователей. Подсчитаем сумму средних чеков.

```python
datesGroups = df_o[['date','group']].drop_duplicates()
ordersAggregated = datesGroups.apply(lambda x: df_o[np.logical_and(df_o['date'] <= x['date'], df_o['group'] == x['group'])]\
.agg({'date' : 'max', 'group' : 'max', 'transaction_id' : 'nunique', 'visitor_id' : 'nunique', 'revenue' : 'sum'}), axis=1).sort_values(by=['date','group'])
```

Получим агрегированные кумулятивные по дням данные о посетителях интернет-магазина: 

```python
visitorsAggregated = datesGroups.apply(lambda x: df_v[np.logical_and(df_v['date'] <= x['date'], df_v['group'] == x['group'])].agg({'date' : 'max', 'group' : 'max', 'visitors' : 'sum'}), axis=1).sort_values(by=['date','group']) 
```

Объединим обе таблицы в одну

```python
cumulativeData = ordersAggregated.merge(visitorsAggregated, left_on=['date', 'group'], right_on=['date', 'group'])
cumulativeData.columns = ['date', 'group', 'orders', 'buyers', 'revenue', 'visitors']

print(cumulativeData.head(5)) 
```

Небольшой фактчекинг, проверим минимальные и максимальные даты в первоначальных данных и в кумулятивных.

```python
df_o[['date']].agg({'date':['min', 'max']})
```

## 6. Проверка статистически значимых результатов.

#### 6.1 Расчёт статистической значимости различий в среднем количестве заказов на посетителя между группами (до удаления выбросов):

***

- **H0**: Нулевая гипотеза предполагает, что две совокупности (A и B) не имеют статистически значимых различий по показателю среднего количества заказов на пользователя.
- **H1**: Альтернативная гипотеза предполагает, что две совокупности (A и B) имеют статистически значимые различия по показателю среднего количества заказов на пользователя.
- **α** (уровень значимости статистического теста) = 0.05
- Для расчёта статистической значимости результатов теста будет применён **U-критерий Манна - Уитни**. Распределение не является нормальным.

***

Рассчитаем кумулятивное количество пользователей в выбранную дату в группе A и Б.

```python
visitorsADaily = df_v[df_v['group'] == 'A'][['date', 'visitors']]
visitorsADaily.columns = ['date', 'visitorsPerDateA']
visitorsACummulative = visitorsADaily.apply(
    lambda x: visitorsADaily[visitorsADaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateA': 'sum'}
    ),
    axis=1,
)
visitorsACummulative.columns = ['date', 'visitorsCummulativeA']

visitorsBDaily = df_v[df_v['group'] == 'B'][['date', 'visitors']]
visitorsBDaily.columns = ['date', 'visitorsPerDateB']

visitorsBCummulative = visitorsBDaily.apply(
    lambda x: visitorsBDaily[visitorsBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'visitorsPerDateB': 'sum'}
    ),
    axis=1,
)
visitorsBCummulative.columns = ['date', 'visitorsCummulativeB']
```

Далее подсчитаем суммарное количество заказов до выбранной даты включительно в группе А и B

```python
ordersADaily = (
    df_o[df_o['group'] == 'A'][['date', 'transaction_id', 'visitor_id', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'})
)
ordersADaily.columns = ['date', 'ordersPerDateA', 'revenuePerDateA']

ordersACummulative = ordersADaily.apply(
    lambda x: ordersADaily[ordersADaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateA': 'sum', 'revenuePerDateA': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersACummulative.columns = [
    'date',
    'ordersCummulativeA',
    'revenueCummulativeA',
]

ordersBDaily = (
    df_o[df_o['group'] == 'B'][['date', 'transaction_id', 'visitor_id', 'revenue']]
    .groupby('date', as_index=False)
    .agg({'transaction_id': pd.Series.nunique, 'revenue': 'sum'})
)
ordersBDaily.columns = ['date', 'ordersPerDateB', 'revenuePerDateB']

ordersBCummulative = ordersBDaily.apply(
    lambda x: ordersBDaily[ordersBDaily['date'] <= x['date']].agg(
        {'date': 'max', 'ordersPerDateB': 'sum', 'revenuePerDateB': 'sum'}
    ),
    axis=1,
).sort_values(by=['date'])
ordersBCummulative.columns = [
    'date',
    'ordersCummulativeB',
    'revenueCummulativeB',
]
```

Объединим столбцы с помощью `merge`.

```python
data = (
    ordersADaily.merge(
        ordersBDaily, left_on='date', right_on='date', how='left'
    )
    .merge(ordersACummulative, left_on='date', right_on='date', how='left')
    .merge(ordersBCummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsADaily, left_on='date', right_on='date', how='left')
    .merge(visitorsBDaily, left_on='date', right_on='date', how='left')
    .merge(visitorsACummulative, left_on='date', right_on='date', how='left')
    .merge(visitorsBCummulative, left_on='date', right_on='date', how='left')
)
```

Создадим переменные `ordersByUsersA` и `ordersByUsersB` с каждым покупателем и его количеством заказов.

```python
ordersByUsersA = (
    df_o[df_o['group'] == 'A']
    .groupby('visitor_id', as_index=False)
    .agg({'transaction_id': pd.Series.nunique})
)
ordersByUsersA.columns = ['visitor_id', 'orders']

ordersByUsersB = (
    df_o[df_o['group'] == 'B']
    .groupby('visitor_id', as_index=False)
    .agg({'transaction_id': pd.Series.nunique})
)
ordersByUsersB.columns = ['visitor_id', 'orders'] 
```

Объявим переменные `sampleA` и `sampleB`, где присвоим пользователям без заказов нули в столбце `orders`.

```python
sampleA = pd.concat(
    [
        ordersByUsersA['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleB = pd.concat(
    [
        ordersByUsersB['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)
```

```python
p_value = '{0:.3f}'.format(stats.mannwhitneyu(sampleA, sampleB)[1])
p_value
```

`p_value` значительно меньше `0.05`, а значит, перед нами статистический значимый результат теста и у нас есть основания отклонить нулевую гипотезу. Однако не стоит забывать, что мы пока не трогали выбросы. Теперь высчитаем относительную разницу. 

```python
print("{0:.3f}".format(sampleB.mean() / sampleA.mean() - 1))
```

Как и было замечено ранее по графикам, разница составила порядка 13-14%. Перейдём к расчёту среднего чека. 

#### 6.2 Расчёт статистической значимости различий в среднем чеке заказа между группами (до удаления выбросов):

***

- **H0**: Нулевая гипотеза предполагает, что две совокупности (A и B) не имеют статистически значимых различий по показателю среднего чека покупки.
- **H1**: Альтернативная гипотеза предполагает, что две совокупности (A и B) имеют статистически значимые различия по показателю среднего среднего чека покупки.
- **α** (уровень значимости статистического теста) = 0.05
- Для расчёта статистической значимости результатов теста будет применён **U-критерий Манна - Уитни**. Распределение не является нормальным.

***

Чтобы рассчитать статистическую значимость различий в среднем чеке, передадим критерию `mannwhitneyu()` данные о выручке с заказов. 

```python
p_value = '{0:.3f}'.format(stats.mannwhitneyu(df_o[df_o['group']=='A']['revenue'], df_o[df_o['group']=='B']['revenue'])[1])
p_value
```

`p_value` значительно больше порога в `0.05`, получить статистически значимый результат на "сырых" данных не вышло. Оснований для отклонения нулевой гипотезы нет. Выясним относительную разницу.

```python
print('{0:.3f}'.format(df_o[df_o['group']=='B']['revenue'].mean()/df_o[df_o['group']=='A']['revenue'].mean()-1)) 
```

Относительная разница составляет порядка 26%, но при этом статистически значимый результат не получен, что явно намекает нам на "аномальную" природу возникновения таких различий. Вероятнее всего, именно в группе `B` находился тот самый одиночный заказ на 1,25 млн у.е.

Пора приступить к очистке данных от выбросов и проверить, какие расчёты мы получим по ним.

Узнаем, сколько всего аномальных пользователей:

```python
usersWithManyOrders = pd.concat(
    [
        ordersByUsersA[ordersByUsersA['orders'] > np.percentile(ordersByUsers['orders'], 97)]['visitor_id'],
        ordersByUsersB[ordersByUsersB['orders'] > np.percentile(ordersByUsers['orders'], 97)]['visitor_id'],
    ],
    axis=0,
)
usersWithExpensiveOrders = df_o[df_o['revenue'] > np.percentile(df_o['revenue'], 98)]['visitor_id']
abnormalUsers = (
    pd.concat([usersWithManyOrders, usersWithExpensiveOrders], axis=0)
    .drop_duplicates()
    .sort_values()
)
print(abnormalUsers.head(5))
print(abnormalUsers.shape[0]) 
```

Всего 40 аномальных пользователей. Уберём их из данных и посмотрим, как они повлияли на результаты:

```python
sampleAFiltered = pd.concat(
    [
        ordersByUsersA[
            np.logical_not(ordersByUsersA['visitor_id'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateA'].sum() - len(ordersByUsersA['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
)

sampleBFiltered = pd.concat(
    [
        ordersByUsersB[
            np.logical_not(ordersByUsersB['visitor_id'].isin(abnormalUsers))
        ]['orders'],
        pd.Series(
            0,
            index=np.arange(
                data['visitorsPerDateB'].sum() - len(ordersByUsersB['orders'])
            ),
            name='orders',
        ),
    ],
    axis=0,
) 
```

#### 6.3 Расчёт статистической значимости различий в среднем количестве заказов на посетителя между группами (после удаления выбросов):

***

- **H0**: Нулевая гипотеза предполагает, что две совокупности (A и B) не имеют статистически значимых различий по показателю среднего количества заказов на пользователя.
- **H1**: Альтернативная гипотеза предполагает, что две совокупности (A и B) имеют статистически значимые различия по показателю среднего количества заказов на пользователя.
- **α** (уровень значимости статистического теста) = 0.05
- Для расчёта статистической значимости результатов теста будет применён **U-критерий Манна - Уитни**. Распределение не является нормальным.

***

Применим статистический критерий Манна-Уитни к обработанным данным.

```python
p_value = '{0:.3f}'.format(stats.mannwhitneyu(sampleAFiltered, sampleBFiltered)[1])
p_value
```

Результаты стали ещё более статистически значимыми после удаления выбросов, есть все основания отвергать нулевую гипотезу. Посмотрим, как обстоят дела с относительной разницей в среднем количестве заказов на посетителя между группами.

```python
print('{0:.3f}'.format(sampleBFiltered.mean()/sampleAFiltered.mean()-1)) 
```

Результаты улучшились. Таким образом, мы получили статистически значимый результат в группе `B`, позволяющий увеличить среднее количество заказов для одного посетителя на `18,5%`.

#### 6.4 Расчёт статистической значимости различий в среднем чеке заказа между группами (после удаления выбросов):

***

- **H0**: Нулевая гипотеза предполагает, что две совокупности (A и B) не имеют статистически значимых различий по показателю среднего чека покупки.
- **H1**: Альтернативная гипотеза предполагает, что две совокупности (A и B) имеют статистически значимые различия по показателю среднего среднего чека покупки.
- **α** (уровень значимости статистического теста) = 0.05
- Для расчёта статистической значимости результатов теста будет применён **U-критерий Манна - Уитни**. Распределение не является нормальным.

***

```python
print(
    '{0:.3f}'.format(
        stats.mannwhitneyu(
            df_o[
                np.logical_and(
                    df_o['group'] == 'A',
                    np.logical_not(df_o['visitor_id'].isin(abnormalUsers)),
                )
            ]['revenue'],
            df_o[
                np.logical_and(
                    df_o['group'] == 'B',
                    np.logical_not(df_o['visitor_id'].isin(abnormalUsers)),
                )
            ]['revenue'],
        )[1]
    )
)

print(
    "{0:.3f}".format(
        df_o[
            np.logical_and(
                df_o['group'] == 'B',
                np.logical_not(df_o['visitor_id'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        / df_o[
            np.logical_and(
                df_o['group'] == 'A',
                np.logical_not(df_o['visitor_id'].isin(abnormalUsers)),
            )
        ]['revenue'].mean()
        - 1
    )
) 
```

После обработки данных от выбросов, `p_value` значительно выросло, подобравшись в плотную к отметке в единицу, что означает почти 100% вероятность получить подобную разницу случайным образом. Как и предполагалось, после удаления выбросов, относительная разница между средними чеками в группах опустилась до `3,7%`. Оснований отвергать нулевую гипотезу нет.
